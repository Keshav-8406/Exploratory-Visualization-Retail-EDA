# Dataset Cell

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

np.random.seed(0)
n = 200

df = pd.DataFrame({
    'age': np.random.randint(18, 60, n),
    'income': np.random.randint(20000, 120000, n),
    'spend': np.random.randint(500, 8000, n),
    'category': np.random.choice(['Electronics', 'Clothing', 'Groceries', 'Sports'], n)
})

df.head()
```

**Task 1 — Bar Chart:  Average Spend by Category**



```python
#Calculate average spend by category
avg_spend = df.groupby('category')['spend'].mean()

#Plot Bar Chart
plt.figure()
avg_spend.plot(kind='bar')

#Add Title And Labels
plt.title('Average Spend by Category')
plt.xlabel('Category')
plt.ylabel('Average Spend')

#Show Plot
plt.show()
```

*Explanation*

In this task, I first grouped the data by category and calculated the average spend for each category. Then I used a bar chart to visualize the comparison between categories. This helps in easily identifying which category has higher or lower average spending.


*Observation*

The Electronics category has the highest average spend compared to other categories.



**Task 2 — Scatter Plot: Income vs Spend**

```python
# Create Scatter Plot Using Seaborn
plt.figure()
sns.scatterplot(x='income', y='spend', data=df)

# Add Title And Labels
plt.title('Income vs Spend')
plt.xlabel('Income')
plt.ylabel('Spend')

# Show Plot
plt.show()
```

*Explanation*

In this task, I created a scatter plot using seaborn to understand the relationship between income and spending. Income is taken on the x-axis and spend on the y-axis.

*Observation*
   
   There is a slight positive relationship between income and spending, but it is not very strong.

**Task 3 — Interactive Scatter Plot with Plotly**

```python
#Interactive Scatter Plot
fig = px.scatter(df, x='income', y='spend', color='category',
                 title='Interactive Income vs Spend by Category')

#Show Plot
fig.show()
```

*Explanation*

In this task, I used plotly to create an interactive scatter plot of income vs spend and colored it by category. This helps in better visualization of different categories.

*Observation*

The interactive plot allows us to hover over data points and see detailed information, which is not possible in static plots.
